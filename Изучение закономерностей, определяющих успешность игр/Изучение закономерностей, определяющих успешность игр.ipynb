# Изучение закономерностей, определяющих успешность игр

# Описание проекта
Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Цель исследования:**
1. Выявить факторы, влияющие на успешность игры;
2. Определить потенциально популярный продукт для дальнейшего планирования рекламной кампании.

**Ход исследования:**
1. Изучение общей информации;
2. Предобработка данных;
3. Исследовательский анализ данных:
   - анализ динамики выпускаемых игр;
   - сравнение динамики продаж по платформам;
   - выявление влияния на продажи внутри одной популярной платформы отзывов пользователей и критиков;
   - анализ общего распределения игр по жанрам.
4. Составление портрета пользователя каждого региона
5. Проверка гипотез:
   - средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
   - средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
6. Написание общего вывода

## Предобработка данных

```python
# проверим наличие пропусков в столбцах
data.isna().sum()
```

```python
# Проверим также наличие пропусков в % для более полного понимания.
pd.DataFrame(round(data.isna().mean()*100,)).style.background_gradient('coolwarm')
```

```python
# В столбцах названия игры и жанра есть пропуски в двух строках, заполнить эти пробелы нет никакой возможности, 
# поэтому удалим пропущенные значения
data = data.dropna(subset=['Name'])
data = data.dropna(subset=['Genre'])
```

```python
# изменим названия столбцов, приведем их к единому формату
data.columns = data.columns.str.lower()
```

```python
# Проверим наличие явных дубликатов
data.duplicated().sum()
```

В столбце год выпуска пропусков 2%. Заменить эти данные или заполнить не представляется возможным, поэтому удалим пропущенные значения и заменим данные на целочисленные значения

```python
data = data.dropna(subset=['year_of_release'])
data['year_of_release'] = data['year_of_release'].astype('Int64')
```

```python
data['critic_score'].unique()
```

Как мы видим из уникальных значений в столбце оценка критиков также нужно изменение типа на целочисленное, однако мы этого сделать не можем так как присутствуют пропуски. Убрать их у нас нет возможности.

```python
data['user_score'].unique()
```

Среди значений оценки пользователей присутствуют пропуски значений, кроме того встречается аббревиатура 'tbd'. TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято. Для нашего анализа мы заменим 'tbd' на пропуски. Кроме того, для дальнейших расчетов мы заменим тип данных на числовой.

```python
data['user_score'] = data['user_score'].replace('tbd', np.NaN)
data['user_score'] = data['user_score'].astype('float')
```

```python
data['rating'].unique()
```

Мы видим, что в датасета присутствуют игры со следующими рейтингами от организации ESRB:
- E («Everyone») — «Для всех»
- M («Mature») — «Для взрослых»
- T («Teen») — «Подросткам»
- E10+ («Everyone 10 and older») — «Для всех от 10 лет и старше»
- K-A («Kids to Adults») — «Для всех» использовался только до 1998 года, позже переименован в рейтинг E, поэтому мы его также его заменим на Е
- AO («Adults Only 18+») — «Только для взрослых»
- ЕС («Early childhood») - «Раннее детство»
- RP («Rating Pending») — «Рейтинг ожидается»

Итак, в столбце рейтинг заменим рейтинг K-A на E, пропуски заменим на Unknown, чтобы была возможность при необходимости рассмотреть их отдельно, также заменим RP на Unknown

```python
data.loc[data['rating'].isin(['K-A']), 'rating'] = 'E'
data['rating'] = data['rating'].fillna('Unknown')
data.loc[data['rating'].isin(['RP']), 'rating'] = 'Unknown'
```

```python
data['rating'].unique()
```

```python
data['rating'].value_counts()
```

```python
data['rating'].hist()
```

Как мы видим большая доля игр не имеет рейтинга, игр с рейтингом EC и AO очень мало

Посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец.

```python
data['all_sales'] = data[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis='columns')
data.head()
```

Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков

```python
most_sales_platforms = ['PS3', 'X360', 'PS4', 'XOne', 'Wii']
```

```python
for name_platform in most_sales_platforms:
    x_critic = actual_data[actual_data['platform']==name_platform]['critic_score']
    y_critic = actual_data[actual_data['platform']==name_platform]['all_sales']
    
    x_user = actual_data[actual_data['platform']==name_platform]['user_score']
    y_user = actual_data[actual_data['platform']==name_platform]['all_sales']
    
    
    fig = plt.figure(figsize=(14, 4))
    
    ax1 = fig.add_subplot(121)
    ax1.scatter(x_critic, y_critic)
    ax1.set_title('Зависимость продаж от оценок критиков\n на платформе '+name_platform, fontsize=14)
    ax1.set_xlabel('Оценки критиков', fontsize=14)
    ax1.set_ylabel('Продажи', fontsize=14)
    
    ax2 = fig.add_subplot(122)
    ax2.scatter(x_user, y_user)
    ax2.set_title('Зависимость продаж от оценок пользователей\n на платформе '+name_platform, fontsize=14)
    ax2.set_xlabel('Оценки пользователей', fontsize=14)
    ax2.set_ylabel('Продажи', fontsize=14)
    
```

```python
for name_platform in most_sales_platforms:    
    correl_critic = actual_data[actual_data['platform']==name_platform]['all_sales'].corr(actual_data[actual_data['platform']==name_platform]['critic_score'])
    correl_user = actual_data[actual_data['platform']==name_platform]['all_sales'].corr(actual_data[actual_data['platform']==name_platform]['user_score'])
    print(name_platform, '   ', 'Корреляция продаж и оценок критиков', round(correl_critic, 3), '   ', 'Корреляция продаж и оценок пользователей', round(correl_user,3))
```

Как мы видим из полученных данных в целом зависимость продаж от оценок пользователей меньше, чем от оценок критиков. Оценки критиков оказывавает примерно равное слабое влияние на количества продаж у всех рассматриваемых платформ (коэффициент корреляции 0.3-0.4).

## Проверка гипотез

Первая гипотеза которую мы проверим: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Для начала сформулируем нулевую и альтернативную гипотезы:

    H0 - средние пользовательские рейтинги платформ Xbox One и PC равны;
    H1 - средний пользовательский рейтинг платформы Xbox One отличается от среднего пользовательского рейтинга платформы PC.

Альтернативная гипотеза двусторонняя, т.к. средний рейтинг может отличаться от как в большую так и в меньшую сторону.

```python
# для начала рассчитаем дисперсии для каждой платформы
xbox_one = actual_data[(actual_data['platform'] == 'XOne') & (~actual_data['user_score'].isna())]['user_score']
print('Средний пользовательский рейтинг платформы Xbox One:', round(xbox_one.mean(),2))
xbox_one_var = np.var(xbox_one, ddof=1)
print('Дисперсия платформы Xbox One:', round(xbox_one_var, 2))
pc = actual_data[(actual_data['platform'] == 'PC') & (~actual_data['user_score'].isna())]['user_score']
print('Средний пользовательский рейтинг платформы PC:', round(pc.mean(),2))
pc_var = np.var(pc, ddof=1)
print('Дисперсия платформы PC:', round(pc_var, 2))
```

Поскольку существует разница между дисперсиями (составляет 0.86),  зададим аргумент equal_var=False.

```python
alpha = 0.05

results = st.ttest_ind(xbox_one, pc, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Т.к. p-значение существенно превысило уровень значимости 5%, то отвергнуть нулевую гипотезу мы не можем. Следовательно у нас нет оснований считать, что средние пользовательские рейтинги платформ Xbox One и PC существенно различаются.

Вторая гипотеза, которую мы проверим: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Сформулируем нулевую и альтернативную гипотезы:

    H0 - средние пользовательские рейтинги жанров Action и Sports равны;
    H1 - средний пользовательский рейтинг жанра Action отличается от среднего пользовательского рейтинга жанра Sports.

Альтернативная гипотеза двусторонняя, т.к. средний рейтинг может отличаться от как в большую так и в меньшую сторону.

```python
# для начала рассчитаем дисперсии для каждой платформы
action = actual_data[(actual_data['genre'] == 'Action') & (~actual_data['user_score'].isna())]['user_score']
print('Cредний пользовательский рейтинг жанра Action:', round(action.mean(),2))
action_var = np.var(action, ddof=1)
print('Дисперсия жанра Action:', round(action_var, 2))
sports = actual_data[(actual_data['genre'] == 'Sports') & (~actual_data['user_score'].isna())]['user_score']
print('Cредний пользовательский рейтинг жанра Sports:', round(sports.mean(),2))
sports_var = np.var(sports, ddof=1)
print('Дисперсия жанра Sports:', round(sports_var, 2))
```

Поскольку существует разница между дисперсиями (составляет 1.13),  зададим аргумент equal_var=False.

```python
alpha = 0.05

results = st.ttest_ind(action, sports, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Исходя из имеющихся данных, на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров Action и Sports в пользу альтернативной гипотезы.
